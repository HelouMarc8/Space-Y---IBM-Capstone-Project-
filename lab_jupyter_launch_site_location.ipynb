```python
from folium.plugins import MarkerCluster

# Create a MarkerCluster object and add it to the map
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# Loop through each launch record in the dataframe
for index, record in spacex_df.iterrows():
    # Extract latitude and longitude
    location = [record['Lat'], record['Long']]
    
    # Determine marker color based on success (class = 1) or failure (class = 0)
    color = 'green' if record['class'] == 1 else 'red'
    
    # Create the marker with icon color
    marker = folium.Marker(
        location=location,
        icon=folium.Icon(color=color, icon='rocket', prefix='fa'),
        popup=f"Launch Site: {record['Launch Site']}<br>"
              f"Flight: {record['Flight Number']}<br>"
              f"Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    
    # Add marker to the cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # Approximate radius of earth in km
    R = 6373.0

    # Convert decimal degrees to radians
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# CCAFS LC-40 launch site center
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Find the radius of the launch site (distance from center to edge)
# You would use MousePosition to find the coordinates of the site boundary
# Example: measuring to the northern edge of the site
site_edge_north_lat = 28.56334
site_edge_north_lon = -80.577356

site_radius = calculate_distance(launch_site_lat, launch_site_lon, 
                                  site_edge_north_lat, site_edge_north_lon)
print(f"Launch site radius (center to edge): {site_radius:.3f} KM")
print(f"Launch site radius (center to edge): {site_radius * 1000:.2f} meters")

# Now calculate distance from center to coastline
coastline_lat = 28.56367
coastline_lon = -80.57163

distance_to_coastline = calculate_distance(launch_site_lat, launch_site_lon, 
                                           coastline_lat, coastline_lon)

# Calculate net distance (from site edge to coastline)
net_distance_to_coastline = distance_to_coastline - site_radius

print(f"\nDistance from site center to coastline: {distance_to_coastline:.3f} KM")
print(f"Distance from site edge to coastline: {net_distance_to_coastline:.3f} KM")
print(f"Distance from site edge to coastline: {net_distance_to_coastline * 1000:.2f} meters")

# You can do similar calculations for other directions
# Example: Eastern edge (towards coastline)
site_edge_east_lat = 28.562302
site_edge_east_lon = -80.57635

distance_to_east_edge = calculate_distance(launch_site_lat, launch_site_lon,
                                           site_edge_east_lat, site_edge_east_lon)
print(f"\nLaunch site radius to eastern edge: {distance_to_east_edge:.3f} KM")
print(f"Launch site radius to eastern edge: {distance_to_east_edge * 1000:.2f} meters")
```
